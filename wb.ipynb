```python
def phi_(u, v, K, gamma, p):
    f_eta_new = -M
    v[-1] = -np.einsum('k,ki->i', weights[:M-1], v[:M-1]) / weights[-1]
    for i in range(M):
        logB = K + u[i][:,None] + v[i][None,:]
        maxLogB = logB.max()
        BsStable =  (np.exp(logB - maxLogB)).sum()
        f_eta_new += weights[i]*(np.log(BsStable) + maxLogB - u[i].dot(p[i]))    
    return f_eta_new * gamma
```

```python
def roundB(x, p, q):
    one = np.ones_like(x[0])
    r = p / x.dot(one)
    r[r>1] = 1.
    F = np.diag(r).dot(x)
    c = q / (x.T).dot(one)
    c[c>1] = 1.
    F = F.dot(np.diag(c))
    err_r = p - F.dot(one)
    err_c = q - (F.T).dot(one)
    return F + np.outer(err_r, err_c) / abs(err_r).dot(one)
```

```python
from numpy import linalg as LA
import scipy as sp
class WassersteinBarycenterSolver:
    def __init__(self, n, m, d, q, alpha=1/3, beta=1/6, e_coeff=5):
        self.n = n
        self.m = m
        self.alpha = alpha # theoretically : 1 / kappa
        self.beta  = beta  # theoretically : 1 / (2 * kappa)
        self.e_coeff = e_coeff # theoretically : 5
        self.d = d
        self.dnorm = LA.norm(d, np.inf)
        self.q = q
        self.eps = 1e-5
    
    def multA(self, x):
        n = self.n
        temp = x.reshape(n,n)
        out = np.concatenate((np.sum(temp,axis=1), np.sum(temp,axis=0)), axis=0)
        return out
    
    def multAt(self, y):
        n = self.n
        out = np.repeat(y[:n],n) + np.tile(y[n:],n)
        return out
    
    def G_x(self, x,p,y):
        n = self.n
        m = self.m
        dnorm = self.dnorm
        result = np.zeros((m, n**2))
        for i in range(m):
            result[i] = self.d + 2 * dnorm * self.multAt(y[i])
        return result / m
    
    def G_p(self, x,p,y):
        n = self.n
        m = self.m
        dnorm = self.dnorm
        return - 2 * dnorm / m * np.sum(y[:,:n], axis=0)
    
    def G_y(self, x,p,y):
        n = self.n
        m = self.m
        dnorm = self.dnorm
        
        result = np.zeros((m, 2*n))
        for i in range(m):
            Ax_i = self.multA(x[i])
            result[i, :n] = p - Ax_i[:n]
            result[i, n:] = self.q[i] - Ax_i[n:]
        return 2. * dnorm / m * result
    
    def alternating_minimization(self, v, v_mp1,  u, iters=25):
        # Solve minimization problem of type 
        # min H(x,y) = <v,x> + <u, y> + r(x,y)
        n = self.n
        m = self.m
        dnorm = self.dnorm
        e_coeff = self.e_coeff
        
        x = 1/n**2 * np.ones((m, n**2))
        p = 1/n * np.ones(n)
        y = np.zeros((m, 2*n))
        out_x = x.copy()
        out_p = p.copy()
        out_y = y.copy()
        
        for t in range(iters):
            for i in range(m):
                gamma = 1/(2*e_coeff) * (m / (2 * dnorm) * v[i] + self.multAt(y[i]**2))
                x[i] = sp.special.softmax(-gamma)/np.sum(sp.special.softmax(-gamma))
            gamma = 1/e_coeff * (1 / (2 * dnorm) * v_mp1 + 1/m * np.sum(y[:,:n], axis=0)) 
            p = sp.special.softmax(-gamma)/np.sum(sp.special.softmax(-gamma))
            for i in range(m):
                Ax_i = self.multA(x[i])
                y[i,:n] = -m/(4 * dnorm) * u[i,:n] / (Ax_i[:n] + p + 1e-9)
                y[i,n:] = -m/(4 * dnorm) * u[i,n:] / (Ax_i[n:] + 1e-9)
                mask = np.abs(y[i]) > 1
                y[i, mask] /= np.abs(y[i, mask])
                
            if (np.linalg.norm(out_x - x, ord=1) + np.linalg.norm(out_p - p, ord=1) < self.eps) and (np.linalg.norm(out_y - y, ord=np.infty) < self.eps):
                return x, p, y

            out_x = x
            out_p = p
            out_y = y

        print('AM Failed')
        return out_x,out_p,out_y
        
    def solve(self, max_time, verbose=False):
        n = self.n
        m = self.m
        dnorm = self.dnorm
        alpha = self.alpha
        beta = self.beta
        e_coeff = self.e_coeff
        
        

        # W = 0
        # for i in range(self.q.shape[0]): 
        #     W += ot.emd2(trueBarycenter,self.q[i],C) # exact linear program
        # W = W/self.q.shape[0]


        gamma = eps/4/np.log(n)
        BStable = -C/gamma
        BStable /= BStable.sum()
        q = np.einsum('ij->j', BStable)            
        stopTime = time.time()

        fprimal = 0
        for j in range(M):
            X = roundB(BStable, self.q[j], q)
            fprimal+=(X*C).sum()

        Wc = 0
        for j in range(self.q.shape[0]): 
            Wc += ot.emd2(q , self.q[j], C)
        Wc = Wc/self.q.shape[0]

        

        # neptune.log_metric('Observed accuracy', y=0, x=0)            
        neptune.log_metric('Wasserstein distance', y=Wc, x=0)            
        neptune.log_metric('Primal function', y=fprimal, x=0)

        s_x = np.zeros((m, n**2))
        for j in range(M):
            s_x[j] = BStable.reshape(n**2)
        s_p = np.zeros(n)
        s_y = np.zeros((m, 2*n))
        
        out_x = s_x.copy()
        out_p = s_p.copy()
        out_y = s_y.copy()
        
        r_grad_x = 4 * e_coeff * dnorm / m * (2*np.log(n) + 1) * np.ones((m, n**2))
        r_grad_p = 2 * e_coeff * dnorm * (np.log(n) + 1) * np.ones(n)
        r_grad_y = np.zeros((m, 2*n))
            
        
        lCount=0
        k=0
        wastedTime=0
        startTime = time.time()
        while True:
            v =     s_x - r_grad_x
            v_mp1 = s_p - r_grad_p
            u =     s_y - r_grad_y
    
            z_x, z_p, z_y = self.alternating_minimization(v, v_mp1, u)
            
            v     += alpha * self.G_x(z_x, z_p, z_y)
            v_mp1 += alpha * self.G_p(z_x, z_p, z_y)
            u     += alpha * self.G_y(z_x, z_p, z_y)
            
            w_x, w_p, w_y = self.alternating_minimization(v, v_mp1, u)
            s_x += beta * self.G_x(w_x, w_p, w_y)
            s_p += beta * self.G_p(w_x, w_p, w_y)
            s_y += beta * self.G_y(w_x, w_p, w_y)
        
            out_x = k/(k+1) * out_x + 1/(k+1) * w_x
            out_p = k/(k+1) * out_p + 1/(k+1) * w_p
            out_y = k/(k+1) * out_y + 1/(k+1) * w_y
            
            
            if  time.time() - startTime - wastedTime > lCount:
                stopTime = time.time()
                
                primal = out_x.reshape(m,n,n)

                oAccuracy=0
                for i in range(m):
                    Ax_i = self.multA(primal[i])
                    oAccuracy += abs(out_p - Ax_i[:n]).sum()
                # BTone = np.einsum('kij->kj', primal)            
                # # BTone = np.einsum('kj,k->kj', BTone, 1/BsStable)
                # qhat = np.einsum('k,kj->j', weights, BTone)
                # oAccuracy = np.einsum('k, kj->', weights, abs(BTone - qhat))

                if lCount % 10==0:
                    np.savetxt(str(lCount)+'AREAbar.csv', out_p)
                    neptune.log_artifact(str(lCount)+'AREAbar.csv')


                Wc = 0
                for j in range(self.q.shape[0]): 
                    Wc += ot.emd2(out_p , self.q[j], C)
                Wc = Wc/self.q.shape[0]

                
                fprimal = 0
                for j in range(M):
                    X = roundB(primal[j], self.q[j], out_p)
                    fprimal+=(X*C).sum()
            
                lCount-=-1
                neptune.log_metric('Observed accuracy', y=oAccuracy, x=stopTime - startTime)  
                neptune.log_metric('Wasserstein distance', y=Wc, x=stopTime - startTime)
                neptune.log_metric('Primal function', y=fprimal, x=stopTime - startTime)
                

                startTime += time.time() - stopTime
                
            
            k-=-1
            if time.time() - startTime > maxTime:
                return time.time() - startTime

            
            
solver = WassersteinBarycenterSolver(n, M, C.flatten(), p_ref, M/3, M/(2*3), 5)
t1 = solver.solve(maxTime, verbose=False)
      
            
```

```python

def agm(eps, pref, maxTime):
    n = len(pref[0])
    m = int(np.sqrt(n))    
    # C =  C_(m)
    
    assert True, 'wrong gamma'
    gamma = eps/4/np.log(n)
    K = -C/gamma

    epsp = eps/8
    p = (1-epsp/8)*pref + epsp/(8*n)
    
    mu=0 #ONLY!
    A = 0.
    tau = 1.
    v = np.zeros([2*M,n], np.float64)
    x = np.zeros([2*M,n], np.float64)

    primal = np.zeros([M, n, n], np.float64)
    BStable = np.zeros([M, n, n], np.float64)
    maxLogB = np.zeros([M], np.float64)
    BsStable = np.zeros([M], np.float64)
    gradPhi = np.zeros([2*M-1, n], np.float64)
    uHatStable = np.zeros([M, n], np.float64)
    vHatStable = np.zeros([M, n], np.float64)

    
    f_x = None
    h=np.ones(2, np.float64)
    i=0

    f = lambda x : phi_(x[:M], x[M:], K, gamma, p)

    def agmsdr_iter(i, h, f_x, x, v):
        def check(h, forcereturn=False):
            # print(i,': ',  h)
            y = v + h * (x-v)

            f_y=-M
            allVHatStable = np.ones([n], np.float64)
            y[M:][-1] = -np.einsum('k,ki->i', weights[:M-1], y[M:][:M-1]) / weights[-1]
            for j in range(M):
                BStable[j] = K + y[:M][j][:,None] + y[M:][j][None,:]
#                 logB[j] = (K + np.outer(y[:M][j], one) + np.outer(one, y[M:][j]))
                maxLogB[j] = BStable[j].max()
                BStable[j] -= maxLogB[j]
                
                BStable[j] = np.exp(BStable[j])
                
                uHatStable[j], vHatStable[j] = (BStable[j]).sum(1), (BStable[j]).sum(0)
                BsStable[j] = (uHatStable[j]).sum()
                allVHatStable *= vHatStable[j]**weights[j]
                
                f_y += weights[j]*(np.log(BsStable[j]) + maxLogB[j] - y[:M][j].dot(p[j]))
                gradPhi[:M][j] = gamma*weights[j]*(uHatStable[j]/BsStable[j]-p[j])
                
            f_y *= gamma

            qHat = vHatStable[-1]/BsStable[-1]
            for j in range(M-1):
                gradPhi[M:][j] = gamma*weights[j]*( vHatStable[j]/BsStable[j] - qHat)
            grad_f_y = np.concatenate((gradPhi[:M], gradPhi[M:]), 0)
            if ((grad_f_y*(v[:2*M-1]-y[:2*M-1])).sum() >= 0 and f_x >= f_y) or forcereturn:
                gu = (gradPhi[:M]*gradPhi[:M]).sum()
                gv = (gradPhi[M:]*gradPhi[M:]).sum()
                norm2_grad_f_y = gu + gv

                x_new = y.copy()
                f_x_new = -M
                for j in range(M):
                    if gv > gu:
                        vHatStable[j][vHatStable[j]<1e-299] = 1e-299
                        vstep = allVHatStable/vHatStable[j]
                        vstepmax = vstep.max()
                        vstep=vstep/vstepmax
                        x_new[M:][j]+=np.log(vstep)
                        BsNew=(BStable[j]*(vstep[None,:])).sum()
                        maxLogB[j] += np.log(vstepmax)
                    else:
                        uHatStable[j][uHatStable[j]<1e-299] = 1e-299
                        ustep = p[j]/uHatStable[j]
                        ustep = ustep/ustep.max()
                        x_new[:M][j]+=np.log(ustep)
                        BsNew=(BStable[j]*(ustep[:,None])).sum()
                    f_x_new += weights[j]*(np.log(BsNew)+maxLogB[j]-x_new[:M][j].dot(p[j])) 
                f_x_new *= gamma   


                
                return True, (y, f_y, grad_f_y , norm2_grad_f_y, x_new, f_x_new, np.einsum('kij,k->kij', BStable, 1/BsStable), h, gv > gu), None
            else:
                return False, grad_f_y, f_y

        if f_x==None: f_x = f(x) 
        tl=None
        fl=fr=None
        tr=np.float128(h)
        k=0
        while True: #find right endpoint for line search
            is_ok, ret, fval = check(tr)
            if is_ok:
                return ret
            else:
                gr = ret
            if (gr*(x[:2*M-1]-v[:2*M-1])).sum() < 0:
                tl=tr
                fl=fval
                tr= 1 + 1e-8 * k**10
            else:
                fr=fval
                break
            k-=-1

        # small step to the left    
        tmp=max(0, min(tr-(1 - (i+1)/(i+2)), (i+1)/(i+2))) 
        while tl==None: #find left endpoint for line search
            is_ok, ret, fl = check(tmp)
            if is_ok:
                return ret
            else:
                gtmp = ret
            if (gtmp*(x[:2*M-1]-v[:2*M-1])).sum() < 0:
                tl=tmp
                break
            else:
                tr=tmp
            tmp = 1 - (1-tmp**(4)) 
        
        # search in [tl,tr]

        while True:
            if tr < 0.8 or tl >= 1.:
                tc = tl + (tr-tl)*3/5
            else:
                tc = tl + (tr-tl)*4/5

            is_ok, ret, fval = check(tc, tc==tr or tc==tl)

            if is_ok:
                return ret
            else:
                gc = ret
            if (gc*(x[:2*M-1]-v[:2*M-1])).sum() > 0:
                tr=tc
                fr=fval
            else:
                tl=tc
                fl=fval


        tc = tl + 0.5* ( tr - tl )
        is_ok, ret, fc = check(tc)
        if is_ok:
            return ret
        else:
            gc = (ret*(x[:2*M-1]-v[:2*M-1])).sum()

        while  True:
            _r = ( tc - tr ) * ( fc - fl )
            _q = ( tc - tl ) * ( fc - fr )
            _p = ( tc - tl ) * _q - ( tc - tr ) * _r
            _q = 2.0 * ( _q - _r )

            if ( 0.0 < _q ):
                _p = - _p

            _q = abs ( _q )
    
            _d = _p / _q
            _u = tc + _d
    
            if tl < _u and _u < tr and tr - tl < 1e-4:
                # print('\t\tquad ', tl, ' ', tc, ' ', tr)
                is_ok, ret, _fu = check(_u, _u==tr or _u==tl)
                if is_ok:
                    return ret
                else:
                    _gu = (ret*(x[:2*M-1]-v[:2*M-1])).sum()

                if  _u < tc :
                    if gc < 0:
                        tl = _u
                        fl = _fu
                    else:
                        tr = tc
                        fr =fc
                        tc = _u
                        fc = _fu
                        gc = _gu
                else:
                    if gc > 0:
                        tr = _u
                        fr = _fu
                    else:
                        tl = tc
                        fl = fc
                        tc = _u
                        fc = _fu
                        gc = _gu

            else:
                # print('bise ')
                if gc > 0:
                    tr = tc
                    fr = fc
                else:
                    tl = tc
                    fl = fc

                if tr < 0.8 or tl >= 1.:
                    tc = tl + (tr-tl)*3/5
                else:
                    tc = tl + (tr-tl)*4/5

                is_ok, ret, fc = check(tc, tc==tr or tc==tl)

                if is_ok:
                    return ret
                else:
                    gc = (ret*(x[:2*M-1]-v[:2*M-1])).sum() 

    # W = 0
    # for i in range(pref.shape[0]): 
    #     W += ot.emd2(trueBarycenter,pref[i],C) # exact linear program
    # W = W/pref.shape[0]


    q = np.einsum('ij->j', K/K.sum()) 

    fprimal = 0
    for j in range(M):
        X = roundB(K/K.sum(), pref[j], q)
        fprimal+=(X*C).sum()

    Wc = 0
    for j in range(pref.shape[0]): 
        Wc += ot.emd2(pref[j], q, C)
    Wc = Wc/pref.shape[0]


    # neptune.log_metric('Observed accuracy', y=0, x=0)            
    neptune.log_metric('Wasserstein distance', y=Wc, x=0)            
    neptune.log_metric('Primal function', y=fprimal, x=0)
        
       
       
    
    i=0
    lCount=0
    wastedTime=0
    startTime = time.time()
    while True:
        y, f_y, grad_f_y, norm2_grad_f_y, x, f_x, B, h[i%2], vflag = agmsdr_iter(i, h[i%2], f_x, x, v)
        # print('final ', t[i%2],'\n\n')
        g = (f_x - f_y)
        a = norm2_grad_f_y + 2*mu*g
        b = 2*mu*A*g + 2*tau*g - mu*tau*(((v[:2*M-1] - y[:2*M-1])**2).sum())
        c = 2*A*tau*g

        alpha = (-b + np.sqrt(b*b - 4*a*c)) / 2 / a
        
        primal = alpha * B + primal*A
        A = A + alpha
        primal/=A
        
        v[:2*M-1] = tau*v[:2*M-1] + mu*alpha*y[:2*M-1] - alpha * (grad_f_y)
        tau = tau+mu*alpha
        v[:2*M-1]/=tau

        if vflag and time.time() - startTime > lCount :
            
           
            stopTime = time.time()
            
            if lCount % 10==0:
                np.savetxt(str(lCount)+'AGMbar.csv', q)
                neptune.log_artifact(str(lCount)+'AGMbar.csv')

            BTone = np.einsum('kij->kj', primal)            
            q = np.einsum('k,kj->j', weights, BTone)
            oAccuracy = np.einsum('k, kj->', weights, abs(BTone - q))

            Wc = 0
            for j in range(pref.shape[0]): 
                Wc += ot.emd2(q , pref[j], C)
            Wc = Wc/pref.shape[0]

            fprimal = 0
            for j in range(M):
                X = roundB(primal[j], pref[j], q)
                fprimal+=(X*C).sum()

            lCount-=-1
            neptune.log_metric('Observed accuracy', y=oAccuracy, x=stopTime - startTime)  
            neptune.log_metric('Wasserstein distance', y=Wc, x=stopTime - startTime)
            neptune.log_metric('Primal function', y=fprimal, x=stopTime - startTime)
                


            startTime += time.time() - stopTime
            
            
        if time.time() - startTime > maxTime:
            return time.time() - startTime
            
    
        i-=-1
```
